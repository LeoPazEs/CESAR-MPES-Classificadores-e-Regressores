# Importando libs importantes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
```

## Selecionando base de dados

```python
iris = datasets.load_iris()

dados = iris.data
rotulos = iris.target #NÃ£o usaremos
```

```python
iris
```

## Selecionando agrupador

```python
#Finding the optimum number of clusters for k-means classification
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 500, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(dados)
```

```python
#Visualising the clusters
plt.scatter(dados[y_kmeans == 0, 2], dados[y_kmeans == 0, 3], s = 100, c = 'purple', label = 'Roxo')
plt.scatter(dados[y_kmeans == 1, 2], dados[y_kmeans == 1, 3], s = 100, c = 'orange', label = 'Laranja')
plt.scatter(dados[y_kmeans == 2, 2], dados[y_kmeans == 2, 3], s = 100, c = 'green', label = 'Verde')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'red', label = 'Centroids')

plt.legend()
```
