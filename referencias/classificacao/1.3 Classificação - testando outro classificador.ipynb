# Importando libs importantes

```python
from sklearn import datasets
from sklearn import tree
import numpy
```

## Selecionando base de dados

```python
iris = datasets.load_iris()

dados = iris.data
rotulos = iris.target
```

## Selecionando classificador

```python
clf = tree.DecisionTreeClassifier() #Vamos usar um white box
```

## Plotando a Ã¡rvore

```python
#!pip install graphviz
import graphviz

import matplotlib.pyplot as plt

clf = clf.fit(dados, rotulos)

fig, ax = plt.subplots(figsize=(20, 20))
tree.plot_tree(clf, fontsize=14, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

## Vamos usar validaÃ§Ã£o cruzada



```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(clf, dados, rotulos, cv=4)

print(numpy.mean(scores))
```

## Outras mÃ©tricas de avaliaÃ§Ã£o

```python
from sklearn.model_selection import cross_validate

scoring_list = ['precision_macro', 'recall_macro', 'f1_macro', 'accuracy']

scores = cross_validate(clf, dados, rotulos, cv=4, scoring=scoring_list)

print(scores)
```
