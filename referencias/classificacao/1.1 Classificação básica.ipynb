## Selecionando um dataset do sklearn

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()

print("Rotulos:", iris.target,"\n")
print("Dados:", iris.data, "\n")
print("Nomes dos rotulos:", iris.target_names)
```

## Selecionando um algoritmo de classificação

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

## Treinando o algoritmo de classificação

```python
dados = iris.data
rotulos = iris.target

clf.fit(dados, rotulos)
```

## Classificando planta nova

```python
planta_nova = [2,3,4,1]

classe = clf.predict([planta_nova])

print(iris.target_names[classe])
```
